### Логические операторы

Расчет дифференциального уравнения второго порядка для угла $\theta$ маятника, на который действует сила тяжести с трением:
$\theta''(t) + b*\theta'(t) + c*sin(\theta(t)) = 0$, $b$ и $c$ — положительные константы

Данное уравнение можно представить как систему уравнений

$\theta'(t) = \omega(t)$

$\omega'(t) = -b*\omega(t) - c*\sin(\theta(t))$

Пусть $y$ будет вектором $\{\theta, \omega\}$, тогда решение:

````python
from scipy.integrate import odeint

def pend(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt


b = 0.25
c = 5.0
y0 = [np.pi - 0.1, 0.0]
t = np.linspace(0, 10, 101)

sol = odeint(pend, y0, t, args=(b, c))

plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
````

### Лямбда функции

Лямбда-выражения в языке Python представляют небольшие анонимные функции, которые определяются с помощью оператора lambda. Формальное определение лямбда-выражения: `lambda [параметры] : инструкция`

`message = lambda: print("hello")`
 
Вызов `message()` выведет сообщение hello

Здесь лямбда-выражение присваивается переменной message. Это лямбда-выражение не имеет параметров, ничего не возвращает и просто выводит строку "hello" на консоль. И через переменную message мы можем вызвать это лямбда-выражение как обычную функцию. Фактически оно аналогично следующей функции:

`def message(): `
    `print("hello")`

````python
square = lambda n: n * n
 
print(square(4))
print(square(5)) 
````

Хотя лямбда-выражения позволяют немного сократить определения функций, тем не менее они ограничены тем, что они могут выполнять только одну инструкцию. Однако они могут быть довольно удобны в тех случаях, когда необходимо использовать функцию для передачи в качестве параметра или возвращения в другой функции. 

````python
def do_operation(a, b, operation):
    result = operation(a, b)
    print(f"result = {result}")

do_operation(5, 4, lambda a, b: a + b)
do_operation(5, 4, lambda a, b: a * b)
````

````python
def generateFibonachi():
    # дописать
    n = int(input("Скольчко чисел Фибоначи нужно сгенерировать?"))
    
    #проверка что n больше 0
    if n <= 0:
        return []
    
    #инициализируем первые два числа Фибоначи
    fib = [1,1]
    
    #генерим последовательность Фибоначи
    for i in range(2,n):
        nextNumber = fib[-1] + fib[-2]
        fib.append(nextNumber)
        
    #возвращаем первые n чисел Фибоначи    
    return fib[:n]
````

Для тестирования функции используйте файл `myText.txt`, в котором текст стихотворения. Стихотворение на ваш выбор.

````python
# дописать
def readTxtAndPrintLastLines(file, lines):
    try:
        #открываем файл для чтения
        with open(file, 'r', encoding='utf-8') as f:
            allLines = f.readlines()
            
            print("Содержимое файла:")
            for line in allLines:
                print(line, end='')
                
            print()
            
            print(f"\nПоследние {lines} строк:")
            for line in allLines[-lines:]:
                print(line, end='')
                
    except FileNotFoundError:
        print(f"Файл {file} не найден.")
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        
file = "D:\poem.txt"
lines = 5
readTxtAndPrintLastLines(file, lines)
````

Гармонический сигнал задается уравнением:

$s(t) = A * \cos(2 \pi f t + \phi)$,

где 
* $A$ – амплитуда колебания, 
* $f$ - частота сигнала, 
* $\phi$ – начальная фаза.

Амплитудная модуляция - широко известный способ изменения формы сигнала по параметру амплитуды. Из названия очевидно, что амплитуда такого сигнала изменяется во времени. Закон изменения амплитуды - произвольный, на практике часто используется модуляция гармоническим колебанием.

$s(t) = A_c(1 + m \cos(\omega_m t + \phi)) * \cos(\omega_c t)$,

где

* $A_c$ – амплитуда несущего колебания,
* $\omega_c$ – частота несущего сигнала,
* $\omega_m$ – частота модулирующего (информационного) сигнала,
* $\phi$ – начальная фаза модулирующего сигнала,
* $m$ - коэффициент модуляции.

Под термином угловая модуляция понимается модуляция по фазе или по частоте (математически можно представить, что изменяется "угол" тригонометрической функции). Таким образом, сигналы с угловой модуляцией делятся на частотно-модулированные (ЧМ) сигналы и фазо-модулированные (ФМ) сигналы.

При фазовой модуляции значение угла фазы изменяется пропорционально информационному сообщению, при частотной модуляции информационный сигнал управляет частотой несущего колебания. В обоих случаях амплитуда сигнала остается неизменной.

$s(t) = A_c \cos (2 \pi (f_c + m * \cos(2\pi f_m t))t + \phi ) $,

где

* $A_c$ – амплитуда несущего колебания,
* $f_c$ – частота несущего сигнала,
* $f_m$ – частота модулирующего сигнала,
* $\phi$ – начальная фаза модулирующего сигнала,
* $m$ - коэффициент модуляции.

**Внимание!** Все графики должны быть оформлены в соответствии с правилами, названия осей, названия линий на графики, легенда, название графика!

````python
# синусойда
import numpy as np
import matplotlib.pyplot as plt

# Функция для генерации синусоиды
def generate_sine(A, f, phi, t):
    return A * np.sin(2 * np.pi * f * t + phi)

# Временная ось
t = np.linspace(0, 2, 1000)  # От 0 до 2 секунд с 1000 точками

# Параметры сигнала
A = 1.0  # Амплитуда
f = 1.0  # Частота
phi = 0.0  # Начальная фаза

# Параметры для модуляции
Ac = 1.0  # Амплитуда несущего колебания
fc = 10.0  # Частота несущего сигнала
fm = 1.0  # Частота модулирующего сигнала
m = 0.5  # Коэффициент модуляции

# Генерация сигналов для различных параметров
s_sine_A = generate_sine(2.0, f, phi, t)  # Увеличенная амплитуда
s_sine_f = generate_sine(A, 2.0, phi, t)  # Увеличенная частота
s_sine_phi = generate_sine(A, f, np.pi/4, t)  # Измененная фаза

# Построение графиков
plt.figure(figsize=(15, 20))

# Синусоида с различными параметрами
plt.subplot(6, 1, 1)
plt.plot(t, generate_sine(A, f, phi, t), label='Синусоида (A=1, f=1, phi=0)')
plt.plot(t, s_sine_A, label='Синусоида (A=2, f=1, phi=0)')
plt.plot(t, s_sine_f, label='Синусоида (A=1, f=2, phi=0)')
plt.plot(t, s_sine_phi, label='Синусоида (A=1, f=1, phi=pi/4)')
plt.title('Синусоида с различными параметрами')
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.legend()

plt.tight_layout()
plt.show()
````

````python
1. На что влияют различные параметры сигнала?
Амплитуда (A, Ac):

Влияние: Определяет максимальное значение сигнала. Чем больше амплитуда, тем выше пиковые значения сигнала.

Пример: В гармоническом сигнале увеличение амплитуды приведет к увеличению высоты пиков синусоиды.

Частота (f, fc, fm):

Влияние: Определяет скорость изменения сигнала во времени. Чем выше частота, тем быстрее сигнал проходит через свои максимальные и минимальные значения.

Пример: В гармоническом сигнале увеличение частоты приведет к более быстрому колебанию синусоиды.

Начальная фаза (phi):

Влияние: Определяет смещение сигнала по времени. Изменение фазы сдвигает сигнал влево или вправо по оси времени.

Пример: В гармоническом сигнале изменение фазы приведет к сдвигу синусоиды по оси времени.

Коэффициент модуляции (m):

Влияние: Определяет глубину модуляции, влияет на изменение амплитуды или частоты в зависимости от типа модуляции.

Пример: В амплитудной модуляции увеличение коэффициента модуляции приведет к более глубокому изменению амплитуды сигнала.

2. На что влияют различные модуляции сигнала?
Амплитудная модуляция (АМ):

Влияние: Изменяет амплитуду сигнала в соответствии с модулирующим сигналом. Это приводит к изменению высоты пиков сигнала.

Пример: В амплитудной модуляции амплитуда несущего сигнала изменяется в соответствии с амплитудой модулирующего сигнала.

Частотная модуляция (ЧМ):

Влияние: Изменяет частоту сигнала в соответствии с модулирующим сигналом. Это приводит к изменению скорости колебаний сигнала.

Пример: В частотной модуляции частота несущего сигнала изменяется в соответствии с амплитудой модулирующего сигнала.
    Заключение:
Различные параметры сигнала (амплитуда, частота, фаза) влияют на форму и поведение сигнала. 
Модуляции (амплитудная, частотная, фазовая) изменяют характеристики сигнала в соответствии с модулирующим сигналом, 
что позволяет передавать информацию через изменения параметров несущего сигнала.
````
